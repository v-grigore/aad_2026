# Distributions

```python
position_color_map = {
    "Goalkeeper": "#7f7f7f",
    "Defender": "#1f77b4",
    "Midfield": "#c9b800",
    "Attack": "#d62728"
}

def plot_position_distribution(df):
    counts = df["position"].value_counts().reset_index()
    counts.columns = ["position", "count"]
    counts["percent"] = 100 * counts["count"] / counts["count"].sum()

    fig = px.bar(
        counts,
        x="position",
        y="count",
        color="position",
        color_discrete_map=position_color_map,
        title="Distribuția jucătorilor pe poziții",
        text="count",
        custom_data=["percent"]
    )

    fig.update_traces(
        texttemplate="%{y} (%{customdata[0]:.1f}%)",
        textposition="outside",
        hovertemplate=(
            "Position: %{x}<br>"
            "Players: %{y}<br>"
            "Percent: %{customdata[0]:.1f}%<extra></extra>"
        )
    )

    fig.update_layout(
        xaxis_title="Position",
        yaxis_title="Număr jucători",
        showlegend=False
    )

    fig.show()

plot_position_distribution(df)
```

```python
sub_position_color_map = {
    "Goalkeeper": "#7f7f7f",

    "Centre-Back": "#1f77b4",
    "Right-Back": "#4fa3d9",
    "Left-Back": "#7dbce2",

    "Defensive Midfield": "#c9b800",
    "Central Midfield": "#e0cf00",
    "Attacking Midfield": "#f0e442",
    "Right Midfield": "#d8cc3a",
    "Left Midfield": "#f5e86e",

    "Centre-Forward": "#d62728",
    "Second Striker": "#e44d4d",
    "Left Winger": "#ff5a5f",
    "Right Winger": "#ff8a80"
}

def plot_sub_position_distribution(df):
    sub_position_order = [
        "Goalkeeper",

        "Centre-Back", "Right-Back", "Left-Back",

        "Defensive Midfield", "Central Midfield", "Attacking Midfield",
        "Right Midfield", "Left Midfield",

        "Centre-Forward", "Second Striker", "Left Winger", "Right Winger"
    ]

    counts = df["sub_position"].value_counts().reindex(sub_position_order, fill_value=0).reset_index()
    counts.columns = ["sub_position", "count"]
    counts["percent"] = 100 * counts["count"] / counts["count"].sum()

    fig = px.bar(
        counts,
        x="sub_position",
        y="count",
        color="sub_position",
        color_discrete_map=sub_position_color_map,
        category_orders={"sub_position": sub_position_order},
        title="Distribuția jucătorilor pe sub-poziții",
        text="count",
        custom_data=["percent"]
    )

    fig.update_traces(
        texttemplate="%{y} (%{customdata[0]:.1f}%)",
        textposition="outside",
        hovertemplate=(
            "Sub-position: %{x}<br>"
            "Players: %{y}<br>"
            "Percent: %{customdata[0]:.1f}%<extra></extra>"
        )
    )

    fig.update_layout(
        xaxis_title="Sub-position",
        yaxis_title="Număr jucători",
        xaxis_tickangle=-35,
        showlegend=False
    )

    fig.show()

plot_sub_position_distribution(df)
```

```python
def plot_league_distribution_with_logos(df, logo_map, league_color_map=None):
    league_order = [
        "Premier League",
        "LaLiga",
        "Bundesliga",
        "Serie A",
        "Ligue 1",
        "Liga Portugal",
        "Eredivisie",
        "Super Lig"
    ]

    counts = df["league"].value_counts().reindex(league_order, fill_value=0).reset_index()
    counts.columns = ["league", "count"]
    counts["percent"] = 100 * counts["count"] / counts["count"].sum()

    fig = px.bar(
        counts,
        x="league",
        y="count",
        color="league",
        color_discrete_map=league_color_map,
        text="count",
        custom_data=["percent"],
        title="Distribuția jucătorilor pe ligi"
    )

    fig.update_traces(
        texttemplate="%{y} (%{customdata[0]:.1f}%)",
        textposition="outside",
        hovertemplate=(
            "League: %{x}<br>"
            "Players: %{y}<br>"
            "Percent: %{customdata[0]:.1f}%<extra></extra>"
        )
    )

    max_count = counts["count"].max()

    fig.update_yaxes(range=[0, 1.15 * max_count])

    fig.update_layout(
        xaxis_title="League",
        yaxis_title="Număr jucători",
        showlegend=False,
        xaxis_tickangle=-30
    )

    # logo-uri CENTRATE în bare
    for _, row in counts.iterrows():
        league = row["league"]
        count = row["count"]

        if league in logo_map and count > 0:
            fig.add_layout_image(
                dict(
                    source=logo_map[league],   # cale locală, URL sau PIL image
                    x=league,
                    y=count * 0.55,               # centru pe verticală
                    xref="x",
                    yref="y",
                    xanchor="center",
                    yanchor="middle",
                    sizex=0.7,                 # aproape cât bara
                    sizey=0.5 * count,         # ocupă ~60% din înălțimea barei
                    sizing="contain",
                    layer="above",
                    opacity=0.95
                )
            )

    fig.show()
```

```python
logo_map = {
    "Premier League": "logos/premierleague.png",
    "LaLiga": "logos/laliga.png",
    "Bundesliga": "logos/bundesliga.png",
    "Serie A": "logos/seriea.png",
    "Ligue 1": "logos/ligue1.png",
    "Liga Portugal": "logos/ligaportugal.png",
    "Eredivisie": "logos/eredivisie.png",
    "Super Lig": "logos/superlig.png"
}

league_color_map = {
    "Premier League": "#3D195B",
    "LaLiga": "#1E1E1E",
    "Bundesliga": "#D20515",
    "Serie A": "#008FD7",
    "Ligue 1": "#091C3E",
    "Liga Portugal": "#006B3F",
    "Eredivisie": "#FF7A00",
    "Super Lig": "#FFFFFF"
}

plot_league_distribution_with_logos(df, logo_map, league_color_map=league_color_map)
```

# Encoding Club Names

```python
club_value = df.groupby("current_club_name")["market_value_in_eur"].mean()

df["club_avg_value"] = df["current_club_name"].map(club_value)

df["club_avg_value_log"] = np.log1p(df["club_avg_value"])
```

```python
id="loo_encoding"
club_sum = df.groupby("current_club_name")["market_value_in_eur"].transform("sum")
club_count = df.groupby("current_club_name")["market_value_in_eur"].transform("count")

df["club_avg_loo"] = (club_sum - df["market_value_in_eur"]) / (club_count - 1)

df["club_avg_loo_log"] = np.log1p(df["club_avg_loo"])
```

# Correlations

```python
# 🔹 1. Selectează toate coloanele numerice
numeric_df = df.select_dtypes(include=["int64", "float64"]).copy()

# 🔹 2. Elimină manual coloanele irelevante
cols_to_drop = [
    "player_id_players_source",
    "player_id_profiles_source",
    "current_club_id",
    "current_national_team_id",
    "market_value_profiles",
    "log_market_value"
]

numeric_df = numeric_df.drop(columns=cols_to_drop, errors="ignore")

# 🔹 3. Calculează matricea de corelație
corr_matrix = numeric_df.corr()

# 🔹 4. Plot heatmap interactiv
fig = px.imshow(
    corr_matrix,
    text_auto=".2f",
    color_continuous_scale="RdBu_r",
    title="Correlation Matrix"
)

fig.update_layout(width=900, height=800)
fig.show()

# 🔥 5. Corelații cu target-ul (market value)
corr_target = corr_matrix["market_value_in_eur"].sort_values(ascending=False)

print("Correlation with market_value_in_eur:")
print(corr_target)

# 🔹 6. Elimină self-correlation
corr_target = corr_target.drop("market_value_in_eur")

# 🔹 7. Barplot pentru vizualizare
corr_df = corr_target.reset_index()
corr_df.columns = ["feature", "correlation"]

fig = px.bar(
    corr_df,
    x="feature",
    y="correlation",
    title="Correlation with Market Value",
    text="correlation"
)

fig.update_traces(texttemplate="%{y:.2f}", textposition="outside")
fig.update_layout(xaxis_tickangle=-45)

fig.show()
```

# Constants

```python
threshold = 0.95
variance_threshold = 1e-3  # pentru numeric

# 🔹 separă coloanele
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = df.select_dtypes(include=["object"]).columns

# ========================
# NUMERIC
# ========================

print("=== Low variance numeric columns ===")
low_variance_cols = []

for col in numeric_cols:
    var = df[col].var()
    if var < variance_threshold:
        low_variance_cols.append((col, var))

for col, var in low_variance_cols:
    print(f"{col}: variance={var}")

# ========================
# CATEGORICAL
# ========================

print("\n=== Near-constant categorical columns ===")
near_constant_cat = []

for col in categorical_cols:
    value_dist = df[col].value_counts(normalize=True, dropna=False)
    if len(value_dist) > 0:
        top_freq = value_dist.iloc[0]
        if top_freq > threshold:
            near_constant_cat.append((col, top_freq))

for col, freq in near_constant_cat:
    print(f"{col}: {freq:.2%} same value")

# ========================
# CONSTANT (general)
# ========================

constant_cols = [col for col in df.columns if df[col].nunique(dropna=False) == 1]

print("\n=== Constant columns ===")
print(constant_cols)
```

- `club_avg_loo_log`

---

### ❌ Dropped Features

#### 🔗 IDs / Technical
- `player_id_players_source` (used for merging)
- `player_id_profiles_source` (used for merging)

---

#### 🏟️ Club-related (redundant / encoded / leakage)
- `current_club_id` (redundant)
- `current_club_name` (encoded)
- `club_avg_value` (leakage)
- `club_avg_value_log` (leakage)
- `club_avg_loo` (alternative encoding, not used)

---

#### 🌍 Country-related
- `country_of_birth` (redundant)
- `country_of_citizenship` (transformed into `country_grouped`)
- `country_grouped` (encoded)

---

#### 📅 Time-related
- `date_of_birth` (used to compute age)
- `age_bucket` (redundant with numeric age)

---

#### ⚽ Position-related
- `position` (redundant with `sub_position`)
- `sub_position` (encoded)

---

#### 🦶 Other categorical
- `foot` (encoded)
- `league` (encoded)

---

#### 🏆 Misc
- `current_national_team_id` (redundant)
- `market_value_profiles` (redundant)
- `market_value_in_eur` → transformed into `log_market_value` (to reduce skewness)

---

#### 📉 Highly Correlated / Removed for Redundancy
- `appearances` (correlated with `minutes_played`)
- `matches_started` (correlated with `minutes_played`)
- `goals` (correlated with `total_shots`)
- `expected_goals` (correlated with `total_shots`)
- `shots_on_target` (correlated with `total_shots`)
- `assists` (correlated with `expected_assists`)
