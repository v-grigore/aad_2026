# Check for player names that dont appear in both dataframes

```python
dup_names = list(set([
    "Idrissa Gueye", "Juan Rodríguez", "Pablo Felipe", "David López",
    "Pedro Lima", "Wesley", "Vitinha", "Mohamed Bamba", "João Costa",
    "Pedro Santos", "Zé Carlos", "João Mendes", "Serginho",
    "Paulo Victor", "Kiki", "Nico González", "Bernardo", "Otávio"
]))

players_clean = players_filtered[~players_filtered['name'].isin(dup_names)]
profiles_clean = profiles[~profiles['name'].isin(dup_names)]

players_clean['name_norm'] = players_clean['name'].apply(normalize_name)
profiles_clean['name_norm'] = profiles_clean['name'].apply(normalize_name)

names_players = set(players_clean['name_norm'])
names_profiles = set(profiles_clean['name_norm'])

only_in_players = names_players - names_profiles
only_in_profiles = names_profiles - names_players
```

```python
import pandas as pd
import unicodedata

def normalize_name(name):
    if pd.isna(name):
        return None
    name = str(name).strip().lower()
    name = unicodedata.normalize("NFKD", name)
    name = name.encode("ascii", "ignore").decode("utf-8")
    name = " ".join(name.split())
    return name

def extract_last_name(name):
    if pd.isna(name):
        return None
    parts = str(name).strip().split()
    if len(parts) == 0:
        return None
    return parts[-1].lower()

def extract_first_name(name):
    if pd.isna(name):
        return None
    parts = str(name).strip().split()
    if len(parts) == 0:
        return None
    return parts[0].lower()

def normalize_player_position(pos):
    mapping = {
        "Attack": "F",
        "Defender": "D",
        "Goalkeeper": "G",
        "Midfield": "M"
    }
    return mapping.get(pos, None)

def normalize_profile_position(pos):
    mapping = {
        "F": "F",
        "D": "D",
        "G": "G",
        "M": "M"
    }
    return mapping.get(pos, None)

def normalize_profile_league(league):
    mapping = {
        "Super Lig": "TR1",
        "Bundesliga": "L1",
        "Serie A": "IT1",
        "LaLiga": "ES1",
        "Ligue 1": "FR1",
        "Premier League": "GB1",
        "Liga Portugal": "PO1",
        "Eredivisie": "NL1"
    }
    return mapping.get(league, None)

# work on copies
players_tmp = players_clean.copy()
profiles_tmp = profiles_clean.copy()

# normalized names
players_tmp["name_norm"] = players_tmp["name"].apply(normalize_name)
profiles_tmp["name_norm"] = profiles_tmp["name"].apply(normalize_name)

# normalized position groups
players_tmp["pos_group"] = players_tmp["position"].apply(normalize_player_position)
profiles_tmp["pos_group"] = profiles_tmp["position"].apply(normalize_profile_position)

# normalized league
players_tmp["league_norm"] = players_tmp["current_club_domestic_competition_id"]
profiles_tmp["league_norm"] = profiles_tmp["league"].apply(normalize_profile_league)

# ---------------------------
# STEP 1: exact full-name match
# ---------------------------
full_matches = players_tmp.merge(
    profiles_tmp,
    on="name_norm",
    how="inner",
    suffixes=("_players", "_profiles")
)

print("Full-name matches:", len(full_matches))

matched_full_names = set(full_matches["name_norm"])

players_unmatched = players_tmp[
    ~players_tmp["name_norm"].isin(matched_full_names)
].copy()

profiles_unmatched = profiles_tmp[
    ~profiles_tmp["name_norm"].isin(matched_full_names)
].copy()

print("Players unmatched after full-name step:", len(players_unmatched))
print("Profiles unmatched after full-name step:", len(profiles_unmatched))

# ---------------------------
# STEP 2: unique (last_name + position + league) fallback
# ---------------------------
players_unmatched["last_name"] = players_unmatched["name_norm"].apply(extract_last_name)
profiles_unmatched["last_name"] = profiles_unmatched["name_norm"].apply(extract_last_name)

players_unmatched["last_pos_league_key"] = list(zip(
    players_unmatched["last_name"],
    players_unmatched["pos_group"],
    players_unmatched["league_norm"]
))

profiles_unmatched["last_pos_league_key"] = list(zip(
    profiles_unmatched["last_name"],
    profiles_unmatched["pos_group"],
    profiles_unmatched["league_norm"]
))

player_last_key_counts = players_unmatched["last_pos_league_key"].value_counts()
profile_last_key_counts = profiles_unmatched["last_pos_league_key"].value_counts()

unique_last_keys = set(player_last_key_counts[player_last_key_counts == 1].index) & \
                   set(profile_last_key_counts[profile_last_key_counts == 1].index)

last_name_matches = players_unmatched[
    players_unmatched["last_pos_league_key"].isin(unique_last_keys)
].merge(
    profiles_unmatched[
        profiles_unmatched["last_pos_league_key"].isin(unique_last_keys)
    ],
    on="last_pos_league_key",
    how="inner",
    suffixes=("_players", "_profiles")
)

print("Last-name + position + league fallback matches:", len(last_name_matches))

# remove rows matched in step 2 before step 3
matched_player_norms_step2 = set(last_name_matches["name_norm_players"]) if "name_norm_players" in last_name_matches.columns else set()
matched_profile_norms_step2 = set(last_name_matches["name_norm_profiles"]) if "name_norm_profiles" in last_name_matches.columns else set()

players_unmatched_2 = players_unmatched[
    ~players_unmatched["name_norm"].isin(matched_player_norms_step2)
].copy()

profiles_unmatched_2 = profiles_unmatched[
    ~profiles_unmatched["name_norm"].isin(matched_profile_norms_step2)
].copy()

print("Players unmatched after last-name fallback:", len(players_unmatched_2))
print("Profiles unmatched after last-name fallback:", len(profiles_unmatched_2))

# ---------------------------
# STEP 3: unique (first_name + position + league) fallback
# ---------------------------
players_unmatched_2["first_name"] = players_unmatched_2["name_norm"].apply(extract_first_name)
profiles_unmatched_2["first_name"] = profiles_unmatched_2["name_norm"].apply(extract_first_name)

players_unmatched_2["first_pos_league_key"] = list(zip(
    players_unmatched_2["first_name"],
    players_unmatched_2["pos_group"],
    players_unmatched_2["league_norm"]
))

profiles_unmatched_2["first_pos_league_key"] = list(zip(
    profiles_unmatched_2["first_name"],
    profiles_unmatched_2["pos_group"],
    profiles_unmatched_2["league_norm"]
))

player_first_key_counts = players_unmatched_2["first_pos_league_key"].value_counts()
profile_first_key_counts = profiles_unmatched_2["first_pos_league_key"].value_counts()

unique_first_keys = set(player_first_key_counts[player_first_key_counts == 1].index) & \
                    set(profile_first_key_counts[profile_first_key_counts == 1].index)

first_name_matches = players_unmatched_2[
    players_unmatched_2["first_pos_league_key"].isin(unique_first_keys)
].merge(
    profiles_unmatched_2[
        profiles_unmatched_2["first_pos_league_key"].isin(unique_first_keys)
    ],
    on="first_pos_league_key",
    how="inner",
    suffixes=("_players", "_profiles")
)

print("First-name + position + league fallback matches:", len(first_name_matches))

# ---------------------------
# COMBINE RESULTS
# ---------------------------
full_matches["match_type"] = "full_name"
last_name_matches["match_type"] = "last_name_position_league_unique"
first_name_matches["match_type"] = "first_name_position_league_unique"

all_matches = pd.concat(
    [full_matches, last_name_matches, first_name_matches],
    ignore_index=True,
    sort=False
)

print("Total matches:", len(all_matches))

# ---------------------------
# STILL UNMATCHED AFTER ALL 3 STEPS
# ---------------------------
matched_player_norms = set(full_matches["name_norm"])
matched_profile_norms = set(full_matches["name_norm"])

if "name_norm_players" in last_name_matches.columns:
    matched_player_norms |= set(last_name_matches["name_norm_players"])
if "name_norm_profiles" in last_name_matches.columns:
    matched_profile_norms |= set(last_name_matches["name_norm_profiles"])

if "name_norm_players" in first_name_matches.columns:
    matched_player_norms |= set(first_name_matches["name_norm_players"])
if "name_norm_profiles" in first_name_matches.columns:
    matched_profile_norms |= set(first_name_matches["name_norm_profiles"])

players_still_unmatched = players_tmp[
    ~players_tmp["name_norm"].isin(matched_player_norms)
].copy()

profiles_still_unmatched = profiles_tmp[
    ~profiles_tmp["name_norm"].isin(matched_profile_norms)
].copy()

print("Players still unmatched after all steps:", len(players_still_unmatched))
print("Profiles still unmatched after all steps:", len(profiles_still_unmatched))
```

# Write the final dataset

```python
import pandas as pd

# =========================
# 1. Extract matched player/profile IDs
# =========================

full_pairs = full_matches[["player_id_players", "player_id_profiles"]].copy()
last_pairs = last_name_matches[["player_id_players", "player_id_profiles"]].copy()
first_pairs = first_name_matches[["player_id_players", "player_id_profiles"]].copy()

matched_pairs = pd.concat(
    [full_pairs, last_pairs, first_pairs],
    ignore_index=True
).drop_duplicates()

print("Matched pairs:", matched_pairs.shape)


# =========================
# 2. Keep only needed columns
# =========================

players_selected = players_filtered[[
    "player_id",
    "current_club_id",
    "country_of_birth",
    "country_of_citizenship",
    "date_of_birth",
    "sub_position",
    "position",
    "foot",
    "height_in_cm",
    "current_national_team_id",
    "current_club_name",
    "market_value_in_eur"
]].copy()

profiles_selected = profiles[[
    "player_id",
    "name",
    "league",
    "market_value"
]].copy()

stats_selected = stats[[
    "player_id",
    "appearances",
    "matches_started",
    "minutes_played",
    "goals",
    "assists",
    "expected_goals",
    "expected_assists",
    "rating",
    "total_shots",
    "shots_on_target",
    "yellow_cards",
    "red_cards",
    "tackles",
    "interceptions",
    "saves"
]].copy()


# =========================
# 3. Rename IDs so joins are explicit
# =========================

players_selected = players_selected.rename(columns={
    "player_id": "player_id_players"
})

profiles_selected = profiles_selected.rename(columns={
    "player_id": "player_id_profiles",
    "market_value": "market_value_profiles"
})

stats_selected = stats_selected.rename(columns={
    "player_id": "player_id_profiles"
})


# =========================
# 4. Merge matched pairs with players, profiles, stats
# =========================

final_df = matched_pairs.merge(
    players_selected,
    on="player_id_players",
    how="inner"
)

final_df = final_df.merge(
    profiles_selected,
    on="player_id_profiles",
    how="inner"
)

final_df = final_df.merge(
    stats_selected,
    on="player_id_profiles",
    how="left"
)

print("Final dataset shape before market value fill:", final_df.shape)


# =========================
# 5. Check / fill market value
# =========================

# rows where players market value is missing but profiles market value exists
mask_fill_from_profile = (
    final_df["market_value_in_eur"].isna() &
    final_df["market_value_profiles"].notna()
)

n_fill_from_profile = mask_fill_from_profile.sum()
print("Players market value missing, but profiles market value exists:", n_fill_from_profile)

if n_fill_from_profile > 0:
    print("\nRows that will be filled from profiles.market_value:")
    print(
        final_df.loc[
            mask_fill_from_profile,
            ["name", "league", "market_value_in_eur", "market_value_profiles"]
        ].sort_values("name")
    )

# fill players.market_value_in_eur from profiles.market_value when missing
final_df["market_value_in_eur"] = final_df["market_value_in_eur"].fillna(
    final_df["market_value_profiles"]
)

# rows where both are missing
mask_both_missing = (
    final_df["market_value_in_eur"].isna() &
    final_df["market_value_profiles"].isna()
)

n_both_missing = mask_both_missing.sum()
print("\nPlayers with both market values missing:", n_both_missing)

if n_both_missing > 0:
    print("\nRows with both market values missing:")
    print(
        final_df.loc[
            mask_both_missing,
            ["name", "league", "market_value_in_eur", "market_value_profiles"]
        ].sort_values("name")
    )


# =========================
# 6. Optional cleanup
# =========================

final_df = final_df.rename(columns={
    "player_id_players": "player_id_players_source",
    "player_id_profiles": "player_id_profiles_source"
})

final_columns = [
    "player_id_players_source",
    "player_id_profiles_source",
    "name",
    "league",
    "current_club_id",
    "current_club_name",
    "country_of_birth",
    "country_of_citizenship",
    "date_of_birth",
    "sub_position",
    "position",
    "foot",
    "height_in_cm",
    "current_national_team_id",
    "market_value_in_eur",
    "market_value_profiles",
    "appearances",
    "matches_started",
    "minutes_played",
    "goals",
    "assists",
    "expected_goals",
    "expected_assists",
    "rating",
    "total_shots",
    "shots_on_target",
    "yellow_cards",
    "red_cards",
    "tackles",
    "interceptions",
    "saves"
]

final_df = final_df[final_columns]


# =========================
# 7. Save CSV
# =========================

final_df.to_csv("data/final_player_dataset.csv", index=False)
print("\nSaved to data/final_player_dataset.csv")
```
